## 📐 GPR_custom 공식 상세 (F3 채택)

```python
print('=' * 60)
print('GPR_custom 최종 공식 (F3 — 상대적 보도량 × tone 복합)')
print('=' * 60)
print()
print('  F3(t) = −mean_tone(t) × log(1 + N(t) / N̄)')
print()
print('  N(t)  : t일의 이란 지정학 관련 기사 수')
print('  N̄    : 해당 이벤트 기간 전체 평균 기사 수 (이벤트별 독립)')
print('  log   : 자연로그 (numpy.log1p 사용)')
print('  정규화: Z-score (이벤트별 독립)')
print()
print('기존 F3 (log(1+N), 절대량) 대비 개선점:')
print('  - 절대량 → 상대량으로 전환')
print('  - 이벤트 간 보도량 규모 차이 제거 (러-우 vs 호르무즈 동등 비교)')
print('  - 이벤트 내 상대적 보도량 급등을 민감하게 포착')
print()
print('공식 GPR(Caldara & Iacoviello 2022)과 Pearson 상관계수:')
corr = {
    'hormuz_crisis'          : 0.193,
    'soleimani_assassination': 0.535,
    'russia_ukraine_invasion': 0.397,
    'israel_hamas'           : 0.602,
    'israel_iran'            : 0.306,
    'us_israel_iran'         : 0.674,
}
for event, r in corr.items():
    flag = '✅' if r >= 0.3 else '⚠️'
    print(f'  {flag} {event:<30} r = {r:.3f}')
print()
print('선행연구 근거:')
print('  - Caldara & Iacoviello (2022, AER) : 뉴스 빈도 기반 GPR 방법론')
print('  - BlackRock BGRI                  : 빈도 × 감성 조정 곱셈 구조')
print('  - Bondarenko et al. (2023, CEPR)  : 이벤트 특화 GPR')
print('  - Culver et al. (2025, Fed)       : NLP + 감성사전 결합')
```
